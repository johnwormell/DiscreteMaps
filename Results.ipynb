```julia
using Gadfly, Distributions
gcolor(str::String) = Theme(default_color=color(str));
function ddplot(p::Plot,pstr::String)
    Gadfly.display(p);
    Gadfly.draw(PDF("../../AMH1/$(pstr).pdf",16cm,10cm),p);
    nothing
end;

function makelayer!(layerm::Vector{Layer},x::Vector,y::Vector,legstr::String,args...)
    thelayer = layer(
    x=x,y=y,
    color = fill(legstr,length(x)),
    args...
    )
    append!(layerm,thelayer)
    nothing
end;
#makelayer!(layerm::Vector{Layer},f::Function,legstr::String,args...) = makelayer!(layerm,plotgrid,f(plotgrid),legstr,args...);

```

```julia
include("DiscreteMaps.jl")
using DiscreteMaps
```

```julia
M = DiscreteMaps.spdoubling(0.05)
NB = 1000
hgd = [0:NB]/NB
pgd = [0.5:NB-0.5]/NB;
```

```julia
NI = 10^8
xh = DiscreteMaps.iterate(M,NI)[:]
(nothing, xhcts) = hist(xh,hgd)
xhcts *= NB / NI;
```

```julia
spdpl = Gadfly.plot(
    layer(x=pgd,y=xhcts,Geom.bar),
    Guide.xlabel("x"),Guide.ylabel("Ï(x)"),
    Scale.y_continuous(minvalue=0.)
)
ddplot(spdpl,"spdoublingplot");
```

```julia
(rho,r) = DiscreteMaps.spectralacim(M,10)
rhopgd = rho(pgd)
layerm = Layer[]
makelayer!(layerm,pgd,rhopgd[:],"Spectral approximation",Geom.line)
makelayer!(layerm,pgd,xhcts[:],"Simulation and binning",Geom.line)
spdnpl = Gadfly.plot(layerm,
    Guide.colorkey(""),
    Guide.xlabel("x"),Guide.ylabel("Ï(x)"),
    Scale.y_continuous(minvalue=0.)
)
ddplot(spdnpl,"spdoublingnplot");
```

```julia
Ns = [10:10:100,120:20:200,250:50:500]#[21,22,31,32,41,42,51,52,76,77,101,102,151,152,201,202,301,302,401,402]
L = minimum(Ns)
NNs = length(Ns)
rv = Array(Float64,L,NNs)
tv = Array(Float64,NNs)
sv1v = Array(Float64,NNs)
sv2v = Array(Float64,NNs)
for i = 1:NNs
    tstart = tic()
    (nothing, r, S) = DiscreteMaps.spectralacim(M,Ns[i];verbose=false)
    tv[i] = toc()
    sl = length(S) 
    sv1v[i] = S[sl]
    sv2v[i] = S[sl-1]
    rv[:,i] = r[1:L]
end
```

```julia
layerm = Layer[]
for i = [1,2,3,7]
    makelayer!(layerm,Ns,abs(rv[i,:] - rv[i,NNs])[:],"rÌ‚$(i-1)",Geom.point)
end
spdrp = plot(layerm,Scale.x_log2,Scale.y_log2,Guide.colorkey(""),Guide.xlabel("N"),Guide.ylabel("Error"))

ddplot(spdrp,"spderror")
sqrt(var(log2(abs(rv[6,1:NNs-1] - rv[6,NNs])))/var(log2(Ns[1:NNs-1]))) |> println
```

```julia
spdtp = plot(
    x=Ns,y=tv,
    Scale.x_log2,Scale.y_log2,
Guide.xlabel("N"),
Guide.ylabel("Time (s)")
)

ddplot(spdtp,"spdtime");
sqrt(var(log2(tv))/var(log2(Ns))) |> println
```
