```julia
using Gadfly, Distributions
gcolor(str::String) = Theme(default_color=color(str));
function ddplot(p::Plot,pstr::String)
    Gadfly.display(p);
#    Gadfly.draw(PDF("/results/l-$(pstr).pdf",16cm,10cm),p);
    nothing
end;

function makelayer!(layerm::Vector{Layer},x::Vector,y::Vector,legstr::String,args...)
    thelayer = layer(
    x=x,y=y,
    color = fill(legstr,length(x)),
    args...
    )
    append!(layerm,thelayer)
    nothing
end;
#makelayer!(layerm::Vector{Layer},f::Function,legstr::String,args...) = makelayer!(layerm,plotgrid,f(plotgrid),legstr,args...);

```

```julia
include("DiscreteMaps.jl")
using DiscreteMaps
NI = 10^6
M = DiscreteMaps.logistic(3.8)
#@time xh = DiscreteMaps.iterate(M,NI)[:];
```

```julia
NB = 1000
d1 = 0.1805
ds = 0.7695
hgd = d1 + ds*[0:NB]/NB
pgd = d1 + ds*[0.5:NB-0.5]/NB;
```

```julia
#NI = 10^6
#xh = DiscreteMaps.iterate(M,NI)[:]
(nothing, xhcts) = hist(xh,hgd)
```

```julia
spdpl = Gadfly.plot(
    layer(x=pgd,y=xhcts,Geom.bar),
    Guide.xlabel("x"),Guide.ylabel("ρ(x)"),
Scale.y_continuous(minvalue=0.,maxvalue=15)
)
ddplot(spdpl,"spdoublingplot");
```

```julia
(mu,r) = DiscreteMaps.spectralacim(M,100)
rhopgd = rho(pgd)
layerm = Layer[]
makelayer!(layerm,pgd,rhopgd[:],"Spectral approximation",Geom.line)
makelayer!(layerm,pgd,xhcts[:],"Simulation and binning",Geom.line)
spdnpl = Gadfly.plot(layerm,
    Guide.colorkey(""),
    Guide.xlabel("x"),Guide.ylabel("ρ(x)"),
Scale.y_continuous(minvalue=0.,maxvalue=13.)
)
ddplot(spdnpl,"spdoublingnplot");
```

```julia
Ns = [10:4:100,120:20:200]#,250:50:500]#[21,22,31,32,41,42,51,52,76,77,101,102,151,152,201,202,301,302,401,402]
L = minimum(Ns)
NNs = length(Ns)
rv = Array(Float64,L,NNs)
tv = Array(Float64,NNs)
sv1v = Array(Float64,NNs)
sv2v = Array(Float64,NNs)
for i = 1:NNs
    tstart = tic()
    (mu, S) = DiscreteMaps.spectralacim(M,Ns[i];verbose=false)
    tv[i] = toc()
    sl = length(S) 
    sv1v[i] = S[sl]
    sv2v[i] = S[sl-1]
    rv[:,i] = [mu.components[2].mag0, mu.components[1].coeffs[1:L-1]]
    (rhobpgd,nothing) = DiscreteMaps.measureint(hgd,mu)
        rhobpgd *= NI / sum(rhobpgd)
    ccdf(Chisq(NB-1),sum(((xhcts - rhobpgd)).^2 ./ (rhobpgd))) |> println
end
```

```julia
layerm = Layer[]
    makelayer!(layerm,Ns,abs(rv[1,:] - rv[1,NNs])[:],"k̂",Geom.point)

for i = [2,3,4,6,9]
    makelayer!(layerm,Ns,abs(rv[i,:] - rv[i,NNs])[:],"r̂$(i-2)",Geom.point)
end
spdrp = plot(layerm,Scale.x_log2,Scale.y_log2,Guide.colorkey(""),Guide.xlabel("N"),Guide.ylabel("Error"))

ddplot(spdrp,"l-spderror.pdf")
sqrt(var(log2(abs(rv[2,5:8] - rv[2,NNs])))/var(log2(Ns[5:8]))) |> println
```

```julia
spdtp = plot(
    x=Ns,y=tv,
    Scale.x_log2,Scale.y_log2,
Guide.xlabel("N"),
Guide.ylabel("Time (s)")
)

ddplot(spdtp,"spdtime");
sqrt(var(log2(sv1v/sv2v))/var(log2(Ns))) |> println
```
